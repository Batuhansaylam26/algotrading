```python
upper=ta2.volatility.bollinger_hband(aapl_minute_closes["close"],20)
middle=ta2.volatility.bollinger_mavg(aapl_minute_closes["close"],20)
lower=ta2.volatility.bollinger_lband(aapl_minute_closes["close"],20)
```

```python
upperband=pd.DataFrame(upper)
lowerband=pd.DataFrame(lower)
middleband=pd.DataFrame(middle)
bands=pd.concat([upperband,lowerband,middleband],axis=1)
bands.index=aapl_minute_closes.index
```

```python
data["ema"]=ta2.trend.ema_indicator(data["close"],50)
data["signal"]=pd.NA
```

```python
for i in range(len(data)):
    if data.iloc[i,6]<data.iloc[i,3] and data.iloc[i-1,6]>data.iloc[i-1,3]:
        data.iloc[i,7]="buy"
    elif data.iloc[i,6]>data.iloc[i,3] and data.iloc[i-1,6]<data.iloc[i-1,3]:
        data.iloc[i,7]="sell"
if pd.Timestamp(data.index[data["signal"]=="sell"].tolist()[0])<pd.Timestamp(data.index[data["signal"]=="buy"].tolist()[0]):
    data.loc[pd.Timestamp(data.index[data["signal"]=="sell"].tolist()[0]),"signal"]=pd.NA
```

```python

fig=go.Figure()
fig.add_trace(
    go.Scatter(x=data.index,y=data["close"])
)
fig.add_trace(
    go.Scatter(x=data.index,y=data["ema"])
)
for i in data.index[data["signal"]=="sell"].tolist():
    fig.add_vline(x=pd.Timestamp(i),line_width=3,line_dash="dash",line_color="red")
for i in data.index[data["signal"]=="buy"].tolist():
    fig.add_vline(x=pd.Timestamp(i),line_width=3,line_dash="dash",line_color="green")
fig
```
